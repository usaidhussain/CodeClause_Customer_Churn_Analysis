# Overview of the Customer Churn Analysis Dataset

```python
import pandas as pd
import numpy as np
df = pd.read_csv('Customer-Churn.csv')
# Display the first few rows of the DataFrame
print(df.head())

# Check the shape of the DataFrame
print(df.shape)

# Check the data types of columns
print(df.dtypes)
# Count the number of customers who churned
churned_count = df['Churn'].value_counts()['Yes']

# Calculate the churn rate
churn_rate = churned_count / df.shape[0] * 100

print("Churn Rate: {:.2f}%".format(churn_rate))
churn_by_internet = df.groupby('InternetService')['Churn'].value_counts(normalize=True).mul(100).unstack()
print(churn_by_internet)
```

# Bar Graph to demonstrate Churn Rates by Internet Service

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Customer-Churn.csv')

# Calculate the churn rates by internet service
churn_by_internet = df.groupby('InternetService')['Churn'].value_counts(normalize=True).mul(100).unstack()

# Plot the churn rates
churn_by_internet.plot(kind='bar', stacked=True, figsize=(8, 6))

# Set the plot labels and title
plt.xlabel('Internet Service')
plt.ylabel('Churn Rate (%)')
plt.title('Churn Rates by Internet Service')

# Display the plot
plt.show()
```

# Bar Graph to demonstrate Churn Rates by Payment Method

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Customer-Churn.csv')

# Calculate the churn rates by payment method
churn_by_payment = df.groupby('PaymentMethod')['Churn'].value_counts(normalize=True).mul(100).unstack()

# Plot the churn rates
churn_by_payment.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set the plot labels and title
plt.xlabel('Payment Method')
plt.ylabel('Churn Rate (%)')
plt.title('Churn Rates by Payment Method')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

# Pie Chart to demonstrate Churn Rates by Payment Method

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Customer-Churn.csv')

# Calculate the churn rates by payment method
churn_by_payment = df.groupby('PaymentMethod')['Churn'].value_counts(normalize=True).mul(100).unstack()

# Create a pie chart for churn rates by payment method
churn_by_payment['Yes'].plot(kind='pie', autopct='%1.1f%%', figsize=(8, 6))

# Set the plot title
plt.title('Churn Rates by Payment Method')

# Set the aspect ratio to 'equal' to ensure a circular pie chart
plt.axis('equal')

# Display the plot
plt.show()
```

# Count of Customers by Payment Method with Churn Status

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Customer-Churn.csv')

# Plot a count plot for payment method
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='PaymentMethod', hue='Churn')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.title('Count of Customers by Payment Method with Churn Status')
plt.xticks(rotation=45)
plt.legend(title='Churn', loc='upper right')
plt.show()
```

# Distribution of Monthly Charges by Contract with Churn Status

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Customer-Churn.csv')

# Plot a violin plot for contract
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='Contract', y='MonthlyCharges', hue='Churn', split=True)
plt.xlabel('Contract')
plt.ylabel('Monthly Charges')
plt.title('Distribution of Monthly Charges by Contract with Churn Status')
plt.legend(title='Churn', loc='upper right')
plt.show()
```
